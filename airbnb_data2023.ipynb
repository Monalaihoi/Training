## Visualization
Video --> https://www.youtube.com/watch?v=JpSMse3eVVg

#### 1. A basic bar chart of the column "neighbourhood_group"

```python

# Assuming 'neighbourhood_group' is the column in your DataFrame
neighborhood_counts = data['neighbourhood_group'].value_counts()

# Plotting a bar chart
neighborhood_counts.plot(kind='bar', color='skyblue')

# Adding labels and title
plt.xlabel('Neighbourhood Group')
plt.ylabel('Count')
plt.title('Distribution of Neighbourhood Groups')

# Display the plot
plt.show()
```

#### 2. A basic GROUPED bar chart comparing columns neighbourhood_group and number_of_reviews

```python

# Assuming 'neighbourhood_group' and 'number_of_reviews' are columns in your DataFrame
grouped_data = data.groupby('neighbourhood_group')['number_of_reviews'].mean()

# Plotting a grouped bar chart
grouped_data.plot(kind='bar', color='skyblue')

# Adding labels and title
plt.xlabel('Neighbourhood Group')
plt.ylabel('Mean Number of Reviews')
plt.title('Mean Number of Reviews by Neighbourhood Group')

# Display the plot
plt.show()
```

#### 3. A basic STACKED and Pivot bar chart

```python
# Assuming 'neighbourhood_group' and 'number_of_reviews' are columns in your DataFrame
pivot_data = data.pivot_table(index='neighbourhood_group', values='number_of_reviews', aggfunc='sum')

# Plotting a stacked bar chart
pivot_data.plot(kind='bar', stacked=True)

# Adding labels and title
plt.xlabel('Neighbourhood Group')
plt.ylabel('Total Number of Reviews')
plt.title('Total Number of Reviews by Neighbourhood Group')

# Display the plot
plt.show()
```

#### 4. Calculating the mean price

```python
# Grouping by 'neighbourhood_group' and calculating the mean price
average_price_by_neighborhood = data.groupby('neighbourhood_group')['price'].mean()

# Create a bar chart
average_price_by_neighborhood.plot.bar()

# Labeling the axes
plt.xlabel('Neighbourhood')
plt.ylabel('Average Price')

# Show the chart
plt.show()
```
