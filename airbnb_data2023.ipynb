#### 4. Clean a column with "strip" - function


##### Example



```python
data["host_id"].astype(str) #mentioning "astype(str)" to use it as a string function
data["host_name"].str.strip("123._/") # remove all the unwanted marks from the host_name

# lstrip = removes marks from the left side
# rstrip = removes marks from the right side
# string --> cadena de caracetres o dato tipo texto
# str --> string
# strip--> espacios en blanco
# To save the changes add --> data["host_name"] =
data["host_name"] = data["host_name"].str.strip("123._/")
data
```

#### Lisätään uusi columni, jossa on samat tiedot mutta ilman välejä--> strips

```python
# Utiliza strip para eliminar espacios al inicio y al final en 'name'
# y guarda el resultado en una nueva columna 'name_without_spaces'
data['name_without_spaces'] = data['name'].str.strip()
data
```

```python
# Elimina los espacios y guarda en una nueva columna 'name_without_spaces'
data['pöö'] = data['name'].str.replace(' ', '')
data
```

```python
data = data.drop(columns= ['pöö', 'name_without_all_spaces', 'name_without_spaces'])
data
```

#### 5. Removing non-numeric values from a column using "replace"- function

```python
data.dtypes
```

```python
# Cambia el tipo de 'host_id' de int a str
data['host_id'] = data['host_id'].astype(str)
```

```python
data.dtypes
```

```python
# ^= iguals everything exept --> a-zA-Z0-9 --> replace the removed marks with nothing --> ''
data["host_id"].str.replace('[^a-zA-Z0-9]', '', regex=True)
# to save the changes
data["host_id"] = data["host_id"].str.replace('[^a-zA-Z0-9]', '', regex=True)
data
```

```python
df_airbnb=pd.DataFrame(data,
                       columns = ['neighbourhood_group' , 'neighbourhood' , 'room_type' , 'price' ,'minimum_nights','number_of_reviews'])
```

```python

data.info()
```

```python
# Obtener conteo de valores únicos en la columna 'neighbourhood_group'
value_counts = df_airbnb['neighbourhood_group'].value_counts()
# Crear gráfico de barras
value_counts.plot.bar()

# Mostrar el gráfico
plt.show()
```

```python
# Obtener conteo de valores únicos en la columna 'neighbourhood_group'
value_counts = df_airbnb['price'].value_counts()
# Crear gráfico de barras
value_counts.plot.bar()

# Mostrar el gráfico
plt.show()
```

```python

##Exploracion de datos con pandas
# Agrupando por 'neighbourhood' y calculando el precio medio
precio_medio_por_barrio = df_airbnb.groupby('neighbourhood_group')['price'].mean()

# Crear gráfico de barras
precio_medio_por_barrio.plot.bar()

# Etiquetando los ejes
plt.xlabel('Barrio')
plt.ylabel('Precio Medio')

# Mostrar el gráfico
plt.show()
```
