#### 2. A basic GROUPED bar chart comparing columns neighbourhood_group and number_of_reviews

```python

# Assuming 'neighbourhood_group' and 'number_of_reviews' are columns in your DataFrame
grouped_data = data.groupby('neighbourhood_group')['number_of_reviews'].mean()

# Plotting a grouped bar chart
grouped_data.plot(kind='bar', color='skyblue')

# Adding labels and title
plt.xlabel('Neighbourhood Group')
plt.ylabel('Mean Number of Reviews')
plt.title('Mean Number of Reviews by Neighbourhood Group')

# Display the plot
plt.show()
```

#### 3. A basic STACKED and Pivot bar chart

```python
# Assuming 'neighbourhood_group' and 'number_of_reviews' are columns in your DataFrame
pivot_data = data.pivot_table(index='neighbourhood_group', values='number_of_reviews', aggfunc='sum')

# Plotting a stacked bar chart
pivot_data.plot(kind='bar', stacked=True)

# Adding labels and title
plt.xlabel('Neighbourhood Group')
plt.ylabel('Total Number of Reviews')
plt.title('Total Number of Reviews by Neighbourhood Group')

# Display the plot
plt.show()
```

#### 5. Boxplot


#### Comparing prices during 2021 in three different neighbourhood groups

```python
# Assuming 'data' is your DataFrame
data_copy = data.copy()  # Create a copy of the original DataFrame

# Convert 'last_review' to datetime on the copied DataFrame
data_copy['last_review'] = pd.to_datetime(data_copy['last_review'])

# Filter data for the year 2021 and specific neighborhood groups on the copied DataFrame
selected_year = 2021
selected_neighborhoods = ['Centro', 'Latina', 'Retiro']
filtered_data = data_copy[
    (data_copy['last_review'].dt.year == selected_year) & 
    (data_copy['neighbourhood_group'].isin(selected_neighborhoods))]

# Create a boxplot to compare 'price' across 'neighbourhood_group'
plt.figure(figsize=(14, 8))
sb.boxplot(x='neighbourhood_group', y='price', data=filtered_data)
plt.title(f'Comparison of Prices Across Selected Neighbourhood Groups for {selected_year}')
plt.show()
```

#### Using violinplot to see the prices of the apartments in different neighbourhood groups that are less than 500€




```python
# Filter the data (if needed)
neighbourhood_price = data[data.price < 500]

# Set the figure size
plt.figure(figsize=(45, 15))

# Create the violinplot
neighbourhood1 = sb.violinplot(data=neighbourhood_price, x='neighbourhood_group', y='price', hue='neighbourhood_group', palette='husl', legend=False)

# Set the title
neighbourhood1.set_title('Distribution of Prices by Neighbourhood', fontname='Arial', fontsize=20)

# Change the font of X and Y axis labels
plt.xticks(fontname='Arial', fontsize=15)
plt.yticks(fontname='Arial', fontsize=15)

# Adding labels and title
plt.xlabel('Neighbourhood', fontname='Arial', fontsize=20)
plt.ylabel('Price', fontname='Arial', fontsize=20)

# Show the plot
plt.show()
```

#### Visualization of outliers with average prices

```python
average_price = dict(markerfacecolor='blue', markeredgecolor='blue', marker='.') 
# This line defines a dictionary (red_square) with properties for the outliers in the boxplot. 
# In this case, it specifies that the outliers should be represented by blue circles (marker='.') with a blue edge color.

data.boxplot(column='price', by='neighbourhood_group', 
           flierprops=average_price, vert=False, figsize=(20,8)) 
# This line creates a boxplot using the boxplot function from Pandas. 
# It groups the data by the 'neighbourhood_group' column and plots the distribution of prices ('price' column) for each group. 
# The flierprops argument is used to customize the appearance of outliers, and in this case, it applies the properties defined in the red_square dictionary.


plt.xlabel('\nMedian Price\n', fontsize=20)
plt.ylabel('\nDistrict\n', fontsize=20)
# These lines set the labels for the x-axis and y-axis, respectively.

plt.title('\nNeighbourhood prices\n', fontsize=20, fontweight='bold')
# This line sets the title for the plot.

plt.suptitle('')
# This line removes the default super title, which is usually automatically generated when creating subplots but isn't applicable in this case.
```
