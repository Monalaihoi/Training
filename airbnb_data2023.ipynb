```python
df_airbnb=pd.DataFrame(data,
                       columns = ['neighbourhood_group' , 'neighbourhood' , 'room_type' , 'price' ,'minimum_nights','number_of_reviews')
```

```python

data.info()
```

```python

df_airbnb.info()
```

```python
# Obtener conteo de valores únicos en la columna 'neighbourhood_group'
value_counts = df_airbnb['neighbourhood_group'].value_counts()
# Crear gráfico de barras
value_counts.plot.bar()

# Mostrar el gráfico
plt.show()
```

```python
# Obtener conteo de valores únicos en la columna 'neighbourhood_group'
value_counts = df_airbnb['price'].value_counts()
# Crear gráfico de barras
value_counts.plot.bar()

# Mostrar el gráfico
plt.show()
```

```python

##Exploracion de datos con pandas
# Agrupando por 'neighbourhood' y calculando el precio medio
precio_medio_por_barrio = df_airbnb.groupby('neighbourhood_group')['price'].mean()

# Crear gráfico de barras
precio_medio_por_barrio.plot.bar()

# Etiquetando los ejes
plt.xlabel('Barrio')
plt.ylabel('Precio Medio')

# Mostrar el gráfico
plt.show()
```

### Ejercicio de graficas con Matplotlib

```python
# Grafico de lineas con plot
plt.plot(df_airbnb['neighbourhood_group'], df_airbnb['price'])
plt.show()
```

```python
plt.plot(df_airbnb['price'], df_airbnb['room_type'])
plt.show()

precio_medio_por_barrio.plot()

plt.xlabel('room_type')
plt.ylabel('Precio Medio')
```

```python
plt.scatter(df_airbnb['price'], df_airbnb['room_type'])
plt.show()

precio_medio_por_barrio.plot()

plt.xlabel('room_type')
plt.ylabel('Precio Medio')
```

```python
df_airbnb.plot(kind='scatter', x='neighbourhood_group', y='price')
plt.show()
```

```python
# Grafico de barras utilizando "bar"

plt.bar(df_airbnb['room_type'], df_airbnb['price'])
plt.show()
```

```python
# grafico de barras ordenado

df_airbnb_sort=df_airbnb.sort_values('price', ascending=False)
plt.bar(df_airbnb_sort['room_type'], df_airbnb_sort['price'])
plt.show()
```

```python
# b: blue.
# g: green.
# r: red.
# y: yellow.

b= 'blue'
g= 'green'
r= 'red'
y= 'yellow'

plt.bar(df_airbnb_sort['room_type'], df_airbnb_sort['price'],
color=['b', 'g', 'r', 'y'])
plt.show()
```

```python
# y=yellow
plt.bar(df_airbnb_sort['room_type'], df_airbnb_sort['price'],
color=['y'])
plt.show()
```

```python
# Grafico de pie utilizando matplotlib

plt.pie(df_airbnb['minimum_nights'],
labels=df_airbnb['neighbourhood_group'])
plt.show()
```

```python
plt.pie(df_airbnb['minimum_nights'],
labels=df_airbnb['room_type'])
plt.show()
```
