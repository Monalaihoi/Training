## 3. Create a new dataframe with Pandas
A DataFrame is a data structure that organizes data into a 2-dimensional table of rows and columns, much like a spreadsheet.

```python
kids_born =pd.DataFrame({'year_born':[2013, 2014, 2012, 2011, 2021],
                          'name':["Mila", "Sara", "Matias", "Juan Vicente", "Mario"],
                          'mom_name':["Mona", "Mona", "Steph", "Lina", "Maria Angelica"],
                          'country':["Finland", "Finland", "Spain", "Colombia", "Colombia"]})
```

```python
kids_born
```

## Rename columns in dataframe

### Rename one column

```python
kids_born.rename(columns={'name':'kid_name'})
```

### Rename more than one column

### Saving all the changes in dataframe
##### Example:
##### 'name_of_dataframe= the_action( for example rename.columns)`

```python
kids_born = kids_born.rename(columns={'name':'kid_name', 'year_born':'birth_year'})
kids_born
```
