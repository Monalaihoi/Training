### 6. Create a bar chart of kid_age

```python
# Create a bar chart for 'kid_age'
plt.figure(figsize=(10, 6))
plt.bar(kids_born['kid_name'], kids_born['kid_age'])
plt.xlabel('Kid Name')
plt.ylabel('Kid Age')
plt.title('Kid Age Distribution')
plt.show()
```

#### To have the the kid_age in ascending order in bars you have to sort the kid_age

```python
# Assuming 'kids_born' is your DataFrame
kids_born = pd.DataFrame({
    'birth_year': [2013, 2014, 2012, 2011, 2021],
    'kid_name': ['Mila', 'Sara', 'Matias', 'Juan Vicente', 'Mario'],
    'mom_name': ['Mona', 'Mona', 'Steph', 'Lina', 'Maria Angelica'],
    'country': ['Finland', 'Finland', 'Spain', 'Colombia', 'Colombia'],
    'kid_age': [10, 9, 11, 12, 2]
})

# Sort the DataFrame by 'kid_age' in ascending order
kids_born = kids_born.sort_values(by='kid_age')

# Create a bar chart for 'kid_age'
plt.figure(figsize=(10, 6))
plt.bar(kids_born['kid_name'], kids_born['kid_age'])
plt.xlabel('Kid Name')
plt.ylabel('Kid Age')
plt.title('Kid Age Distribution (Ascendant Order)')
plt.show()
```

### More visualizations

#### 1. Histogram of kid_ages

```python
# Here you use Matplotlib = plt
plt.figure(figsize=(10, 6))
plt.hist(kids_born['kid_age'], bins=10, edgecolor='black')
plt.xlabel('Kid Age')
plt.ylabel('Frequency')
plt.title('Distribution of Kid Ages')
plt.show()
```

#### 2. Pie chart of countries

```python
country_counts = kids_born['country'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Kids by Country')
plt.show()
```

### Custom colours

```python
# Count the number of kids per country
country_counts = kids_born['country'].value_counts()

# Define custom colors
custom_colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold', 'lightskyblue']

# Create a pie chart with custom colors
plt.figure(figsize=(8, 8))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=90, colors=custom_colors)
plt.title('Distribution of Kids by Country')
plt.show()
```

#### 3. Display the amount of kids per mom in pink colour

```python
mom_counts = kids_born['mom_name'].value_counts()
# Define custom colors
custom_colors = ['skyblue' if mom != 'Mona' else 'pink' for mom in mom_counts.index]

plt.figure(figsize=(10, 6))
plt.bar(mom_counts.index, mom_counts.values, color='pink')
plt.xlabel('Mom Name')
plt.ylabel('Number of Kids')
plt.title('Number of Kids for Each Mom')
plt.xticks(rotation=45, ha='right')
plt.show()
```

#### 4. Scatter Plot of Birth Year vs. Kid Age:

Explore the relationship between birth year and kid age with the definition of colours

```python
# Define colors based on a condition (e.g., age <= 10)
colors = ['magenta' if age <= 10 else 'purple' for age in kids_born['kid_age']]

plt.figure(figsize=(10, 6))
plt.scatter(kids_born['birth_year'], kids_born['kid_age'], c=colors)
plt.xlabel('Birth Year')
plt.ylabel('Kid Age')
plt.title('Scatter Plot of Birth Year vs. Kid Age')
plt.show()
```

#### 5. Boxplot of Kid Ages by Country:

Compare the distribution of kid ages for different countries.

#### 6. Heatmap

```python
plt.figure(figsize=(12, 8))
heatmap_data = kids_born.pivot_table(index='mom_name', columns='kid_name', values='kid_age')
sb.heatmap(heatmap_data, cmap='coolwarm', annot=True, fmt='g')
plt.title('Heatmap of Kid Ages by Mom and Kid')
plt.show()
```

#### 7. Area plot of kids age and name

```python
plt.figure(figsize=(12, 6))
sorted_data = kids_born.sort_values(by='kid_age')
plt.fill_between(sorted_data['kid_name'], sorted_data['kid_age'], color='yellow', alpha=0.4)
plt.xlabel('Kid Name')
plt.ylabel('Kid Age')
plt.title('Area Plot of Kid Ages')
plt.show()
```

#### 8. 3D Plotting
Creating a 3D plot involves using the Axes3D class from the mpl_toolkits.mplot3d module. 


```python
# Assuming 'kids_born' is your DataFrame
fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(kids_born['birth_year'], kids_born['kid_age'], kids_born.index, c='purple', marker='o')

# Set labels for each axis
ax.set_xlabel('Birth Year')
ax.set_ylabel('Kid Age')
ax.set_zlabel('Kid Index')

# Set title for the plot
ax.set_title('3D Scatter Plot of Birth Year, Kid Age, and Kid Index')

plt.show()
```

#### 9. Pair Plot

```python
sb.pairplot(kids_born, hue='mom_name', markers='o', height=2.5)
plt.suptitle('Pair Plot of Kid Data by Mom')
plt.show()
```

```python
categories = kids_born['kid_name'].unique()
num_categories = len(categories)

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Calculate values for each category
mom_names = kids_born['mom_name'].unique()
for mom_name in mom_names:
    values = kids_born[kids_born['mom_name'] == mom_name]['kid_age'].values
    values = np.concatenate((values, [values[0]]))  # Close the circle
    angles = np.linspace(0, 2 * np.pi, num_categories, endpoint=False)
    
    # Plot the radar chart for each mom
    if len(angles) == len(values):  # Check lengths before plotting
        ax.plot(angles, values, label=mom_name)

# Add labels, title, and legend
ax.set_thetagrids(angles * 180 / np.pi, labels=categories)
ax.set_title('Radar Chart of Kid Ages by Mom')
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))

plt.show()
```

#### 10. Violin Plot

```python
# Assuming 'kids_born' is your DataFrame
plt.figure(figsize=(12, 8))
sb.violinplot(x='mom_name', y='kid_age', data=kids_born, inner='quartile')
plt.title('Violin Plot of Kid Ages by Mom')
plt.xlabel('Mom Name')
plt.ylabel('Kid Age')
plt.show()
```
